# Methodology

The FinBERT model was trained by the authors on a large financial corpus consisting of 4.9 billion tokens. This corpus included corporate reports, earnings call transcripts, and analyst reports. The model was trained using both cased and uncased versions, and the vocabulary was constructed from financial corpora using the SentencePiece library. Multiple variants of FinBERT were trained using either the original BERT code or a new financial vocabulary called FinVocab.

# Implementation

```python
# Sentiment Analysis with FinBERT
from transformers import BertTokenizer, BertForSequenceClassification, pipeline
```

```python
finbert = BertForSequenceClassification.from_pretrained('yiyanghkust/finbert-tone', num_labels=3)
tokenizer = BertTokenizer.from_pretrained('yiyanghkust/finbert-tone')
nlp = pipeline("text-classification", model=finbert, tokenizer=tokenizer)
```

```python
output = nlp(['The company is experiencing robust growth, and our liquidity position is strong.',
              'growth is strong and we have plenty of liquidity.',
              'We are facing a capital shortage and exploring options for additional financing.',
              'there is a shortage of capital, and we need extra financing.',
              'formulation patents might protect Vasotec to a limited extent.',
              'The formulation patents may provide limited protection for Vasotec.'])

print(output)
```
